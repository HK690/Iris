```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the 'iris' DataFrame
iris = pd.read_csv('C:/Users/ASUS/Documents/DS PR/DataSet/Iris.csv')  # Assuming iris data is stored in a CSV file

# Convert float columns to integers
iris['PetalLengthCm'] = iris['PetalLengthCm'].astype(int)
iris['PetalWidthCm'] = iris['PetalWidthCm'].astype(int)
iris['SepalLengthCm'] = iris['SepalLengthCm'].astype(int)
iris['SepalWidthCm'] = iris['SepalWidthCm'].astype(int)

# Create subplots for boxplots
fig, axes = plt.subplots(2, 2, figsize=(16, 8))

# Create boxplots for each column
sns.boxplot(data=iris, y='PetalLengthCm', x='Species', ax=axes[0, 0])
sns.boxplot(data=iris, y='PetalWidthCm', x='Species', ax=axes[0, 1])
sns.boxplot(data=iris, y='SepalLengthCm', x='Species', ax=axes[1, 0])
sns.boxplot(data=iris, y='SepalWidthCm', x='Species', ax=axes[1, 1])

# Display the plots
plt.show()
```
